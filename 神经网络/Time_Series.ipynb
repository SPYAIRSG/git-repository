```python
import pandas as pd
import numpy as np
from 
data = pd.read_csv("train.csv", header=None)
data = np.array(data)
print(data)
```

```python
"""
    构建训练数据，滑动窗口
"""
file_length = 111
linedata_length = 735-56
# 输入时间序列长度、输出时间序列长度
output_length = 56
input_step = output_length*4
# 构建的时间序列训练集
train_data = []
label_data = []
for line in range(0, file_length):
    for col in range(0, linedata_length-input_step):
        train_data.append(data[line:line+1,col:col+input_step])
        label_data.append(data[line:line+1,col+input_step:col+input_step+output_length])
train_data = np.array(train_data).reshape(len(train_data), input_step)
label_data = np.array(label_data).reshape(len(label_data), output_length)
```

```python
"""
    构建测试集
"""
test_data = []
for line in range(0, file_length):
    test_data.append(data[line:line+1,linedata_length-input_step:linedata_length])
test_data = np.array(test_data).reshape(len(test_data), input_step)
test_label = pd.read_csv("test.csv", header=None)
test_label = np.array(test_label)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras import regularizers
model = Sequential()
model.add(Dense(512, input_dim=224, activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(128, activation='relu',kernel_regularizer=regularizers.l2(0.01)))
model.add(Dense(56))
model.compile(loss='mean_squared_error', optimizer='adam')

model.fit(train_data, label_data, epochs=100, batch_size=32)
```

```python
result = model.predict(test_data)
```
